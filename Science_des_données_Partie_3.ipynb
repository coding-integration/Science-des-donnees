## Le nettoyage des données

### Pourquoi nettoyer ?

Pourquoi nettoyons-nous les données ? Parce que les ensembles de données qui nous parviennent peuvent avoir différents formats (avec des règles d'analyse incohérentes), des données invalides, des données manquantes, etc. Par exemple, supposons que vous ayez des données de plusieurs volontaires qui ont noté des observations d'oiseaux avec des horodatages, mais que vous voyez des horodatages formatés de différentes manières entre vos volontaires, comme `15:01`, `15h01`, `15H01`, `1er juin 2023 15h01 EST`, et ainsi de suite. Vous devez les normaliser dans un format unique afin de pouvoir les comparer dans votre analyse. De plus, vous avez peut-être des données numériques pour les températures, mais certaines lignes ne précisent pas si le système utilisait Celsius ou Fahrenheit et doivent être exclues.
<br><br>

Dans le cas de Statistique Canada, l'ensemble de données nous parvient déjà relativement propre, avec des formats bien étiquetés et prévisibles, mais il peut encore y avoir des données manquantes là où il y a des lacunes dans leurs enregistrements (généralement représentées par `-` ou `..` dans leurs tableaux). Il y aura également beaucoup de données superflues, par exemple des métadonnées, des colonnes reliant les données à d'autres enregistrements, ou des colonnes où chaque ligne a la même valeur (par exemple, 'Canada' car nous avons téléchargé un ensemble de toutes les régions géographiques).

<br><br>

### Comment effectuer le nettoyage ?

Il existe essentiellement trois façons de nettoyer les données :

* Manuellement, à l'aide d'un éditeur de texte ou d'un éditeur CSV dédié. Par exemple, VSCode a de nombreuses extensions CSV qui vous permettent d'ajouter ou de réorganiser des colonnes, etc. (Notez que bien que vous puissiez utiliser Excel pour cela, il a tendance à changer le format CSV de manière inattendue.)

* Via un script qui modifie et enregistre le fichier CSV. Par exemple, Python dispose d'un module CSV intégré que nous pourrions utiliser pour lire le fichier ligne par ligne, effectuer des opérations sur les lignes et les colonnes, etc.

* Dans le code pendant l'exécution en utilisant des bibliothèques orientées tableaux comme `pandas`, comme nous l'avons vu dans le notebook précédent.

<br><br>

|Méthode |	Avantages |	Inconvénients
| - | - | -
| Édition manuelle	| Nécessite moins de compétences techniques ; rapide à mettre en place	| Limitée à la recherche et au remplacement ou à l'édition ligne par ligne/colonne par colonne
| Scripting avec le module csv	| Relativement rapide et puissant	| Pas de modèle, donc l'édition des données via les indices plutôt que les en-têtes peut être moins intuitive
| Utilisation de la bibliothèque `pandas` à l'exécution	| Édition puissante et intuitive	| Peut nécessiter une préparation préalable des données ; plus lent, car il est exécuté à chaque exécution

### Nettoyage des données - Partie 2

Commençons par convertir des lignes en colonnes. C'est quelque chose qui peut être fait assez facilement avec la logique Python normale, mais qui serait assez complexe si nous utilisons les `DataFrames` de `pandas`. Nous allons donc utiliser la deuxième méthode.

Notre algorithme lira chaque ligne, identifiera les deux premières colonnes, créera une ligne dans un nouveau fichier CSV, puis ajoutera les colonnes  «Valuer» des lignes successives jusqu'à ce que les deux premières colonnes changent.

Une difficulté mineure que nous rencontrons lorsque en parcourant les données est que toutes les biens n'ont pas le détail complet. Cela signifie que nous devrons lire l'étiquette des données disponibles plutôt que de simplement nous fier à un index. (Statistiques Canada a fourni des données quelque peu «sales» , car la véritable prévisibilité aurait consisté à marquer les données manquantes avec des cellules vides, mais ils ont plutôt omis complètement les lignes. C'est ce qui rend notre travail amusant...!)

Voici le code pour nettoyer le CSV.

````python
# Définir la racine des données
#racine_donnees = 'https://raw.githubusercontent.com/coding-integration/Science-des-donnees/main/donnees/'
racine_donnees = 'donnees'
````

````python
# Deuxième passe du nettoyage des données alimentaires, en utilisant le module csv intégré de Python
import csv

# Ouvrir les anciens et nouveaux fichiers CSV. Nouvelle ligne vide car csv ajoute déjà '\n'
with open(f'{racine_donnees}/aliments_propres_1.csv', 'r') as f_old:
    with open(f'{racine_donnees}/aliments_propres_2.csv', 'w', newline='') as f_new:
        reader = csv.reader(f_old)
        writer = csv.writer(f_new)

        # Omettre les en-têtes ; nous les connaissons
        next(reader)

        # Écrire les en-têtes pour le nouveau fichier
        writer.writerow(['Année', 'Produits', 'Offre totale', 'Stocks initiaux', 'Production', 'Imports', 'Disposition totale', 'Disparition intérieure', 'Exports', 'Déchets', 'Stocks finals'])


        # Modèle de ligne
        # C'est la partie compliquée, où nous devons pouvoir placer chaque valeur dans la bonne colonne
        # Note : Je combine également "Production" et "Fabrication"
        col_indices = {'Offre totale': 2, 'Stocks initiaux': 3, 'Production': 4, 'Fabrication': 4, 'Imports': 5, 'Disposition totale': 6, 'Disparition intérieure': 7, 'Exports': 8, 'Déchets': 9, 'Stocks finals': 10}

        # Commencer à construire la ligne plus fine du CSV
        new_row = [''] * 11

        # Examiner chaque ligne
        for old_row in reader:

            # Écrire une ligne terminée
            if new_row[0] and (old_row[:2] != new_row[:2]):
                writer.writerow(new_row)
                new_row = [''] * 11

            # Construire ladite ligne
            new_row[0] = old_row[0]
            new_row[1] = old_row[1]

            label = old_row[2]
            col_index = col_indices[label]
            value = old_row[3]

            new_row[col_index] = value

        # Après la boucle, écrire la dernière ligne
        writer.writerow(new_row)
````

### Sous-catégories

Un inconvénient que je constate est qu'il y a 190 différentes marchandises et les catégories ne sont pas si évidentes. Cependant, une lecture plus attentive par année suggère que chaque année contient les catégories suivantes : farines, divers, boissons, produits laitiers, viandes, huiles, fruits, légumes, poissons. Nous voulons probablement les regrouper si nous voulons comparer des aliments, sinon nous comparerons du lait écrémé à 1 % avec du lait écrémé à 3 %, par exemple.

Il existe plusieurs façons de le faire. Ma méthode consistera à utiliser davantage de scripts Python pour créer (1) une liste de valeurs uniques et (2) un fichier CSV dans lequel je vais mapper manuellement les valeurs aux catégories. Comme cette carte sera assez petite, je la consulterai au moment de l'analyse des données plutôt que de créer une troisième révision de notre ensemble de données.

Voici du code qui crée la base du mappage des sous-catégories. Il génère `aliments_tige.csv`.

````python
# Créer un mappage de sous-catégorie
import csv

# Utilisez un ensemble pour suivre les produits uniques, car les ensembles résolvent intrinsèquement les doublons
unique_produits = set()

with open(f'{racine_donnees}/population_propres_2.csv', 'r') as f_data:
        reader = csv.reader(f_data)

        # Ignorer les en-têtes
        next(reader)

        # Mettre chaque produit dans l'ensemble
        for row in reader:
            produits = row[1]
            unique_produits.add(produits)

# Écrivez au format CSV pour ajouter manuellement des catégories
with open(f'{racine_donnees}/aliments_tige.csv', 'w', newline='') as f_map:
    writer = csv.writer(f_map)

    # Écrivez les en-têtes
    writer.writerow(['Produits', 'Catégorie'])

    # Écrivez chaque ligne avec un espace vide dans la colonne de catégorie
    for produits in sorted(unique_produits):
         writer.writerow([produits, ''])
````

### Dataframe de multi-niveaux

Bon. Maintenant, supposons que nous voulions créer un `DataFrame` par catégorie. Cela suggère que nos lignes sont des catégories. (Nous éliminerons les matières premières pour ce cadre.) Quelles seront les colonnes ? Supposons que nous disions des années. Si tel est le cas, nous devons choisir l'une des mesures pour servir de vecteur. Par exemple, si notre cadre ressemblait à ceci :

| Catégorie | 1960 | 1961 | ... | 2022
| - | - | - | - | -
| Agrafes | 200.2 | 250.4 | ... | 678.2
| Breuvages | 37.3 | 49.1 | ... | 163.0

Ces chiffres ne peuvent apparemment représenter qu'une de nos mesures précédentes : l'offre totale, ou l'élimination totale, ou peut-être les importations, les exportations ou les déchets.

Quoi alors ? Avons-nous besoin d'une base de données distincte pour chacune des 10 mesures ? On pourrait peut-être le faire avec une fonction :

````python
# Fonction de création d'un dataframe par catégorie pour une colonne donnée

def make_catwise(column: str) -> pd.DataFrame:
    df_cats = pd.DataFrame()
    df_cats['Catégorie'] = df['Catégorie'].unique()
    df_cats = df_cats.set_index('Catégorie', drop=True)

    # Passer chaque année
    for year in df['Année'].unique():
        df_cats[year] = 0.0

        # Additionnez les années de la colonne fournie pour chaque catégorie
        for cat in df_cats.index.get_level_values('Catégorie'):
            data = df[(df['Année'] == year) & (df['Catégorie'] == cat)][column]
            total = sum(data.fillna(0))
            df_cats.loc[cat, year] = total

    return df_cats
````

````python
# Tester ladite fonction
df_cat_supply = make_catwise('Offre totale')
print(df_cat_supply.head())
````

Cependant, il semble peu intuitif d'avoir 10 `DataFrames` distincts qui se concentrent chacun sur une mesure.

Au lieu de cela, nous pouvons utiliser un [`MultiIndex`](https://pandas.pydata.org/docs/user_guide/advanced.html). Ici, les indices sont constitués de plusieurs niveaux hiérarchiques. Si cela semble complexe, voici une illustration de tableau :


| Catégorie | Année | Offre totale | Disposition totale
| - | - | - | -
| Agrafes | 1960 | 200.2 | 196.4
| | 1961 | 250.4 | 212.3
| | ... | ... | ...
| | 2022 | 678.2 | 654.7
| Breuvages | 1960 | 37.3 | 33.4
| | 1961 | 49.1 | 48.9
| | ... | ... | ...
| | 2022 | 163.0 | 155.6

Cette indexation hiérarchique nous permet de saisir chaque année *et* mesure tout en l'orientant autour d'une seule des colonnes d'origine (Catégorie).

Voici comment nous le fabriquons :

````python
# Créer un DataFrame par catégorie avec un MultiIndex

# Définissez le multi-index sur la catégorie et l'année ; laisser tomber la définition des produits de base

df_cats = df.set_index(['Catégorie', 'Année'])
df_cats = df_cats.drop(columns='Produits')

# Somme des années
df_cats = df_cats.groupby(level=[0, 1]).sum(numeric_only=True)

print(df_cats.head())
````

C'est la puissance de `DataFrames` dans `pandas`. Pensez à toutes les itérations et comparaisons que nous aurions faites sur des structures de données Python normales !

Notez qu'en voyant cette logique, il sera facile de revendiquer la spécificité au niveau de la marchandise si nous voulons :

````python
# Créer un DataFrame par produit avec un MultiIndex

# Définissez le multi-index sur le produit et l'année
df_coms = df.set_index(['Catégorie', 'Année'])

# Somme des années
df_coms = df_coms.groupby(level=[0, 1]).sum(numeric_only=True)

print(df_coms.head())
````

###Comment notre production par rapport à nos importations a-t-elle changé en fonction de notre offre totale ?

Essayons un graphique à aires empilées pour représenter la production totale par rapport aux importations totales au fil des années.

````python
#Production vs importations
fig = px.area(df_years_3, x=df_years_3.index, y=['Production', 'Imports'],
              title='Toute la production et l'importation d'aliments au Canada',
              labels={'valeur': 'Tonnes x 1,000'})
fig.show()
````

Une partie de ma question, cependant, était "en fonction de notre offre totale", c'est-à-dire 'partie d'un tout'. Bien que nous puissions voir la croissance relative des deux zones et les comparer visuellement, peut-être devrions-nous les considérer comme des proportions.

````python
# Production vs importations, proportionnellement
prop = df_years_3.drop(['Offre totale', 'Stocks initiaux', 'Disposition totale', 'Disparition intérieure', 'Exports', 'Déchets', 'Stocks finals'], axis=1)
prop['Denominator'] = prop['Production'] + prop['Imports']
prop['Production'] /= prop['Denominator'] / 100
prop['Imports'] /= prop['Denominator'] / 100
prop = prop.drop('Denominator', axis=1)

fig = px.area(prop, x=prop.index, y=prop.columns,
              title='Ensemble de la production et de l'importation d'aliments au Canada (en proportion)',
              labels={'valeur': "% de l'offre"})
fig.show()
````

Ici, nous pouvons voir que la part de l'approvisionnement fournie par les importations a augmenté régulièrement.
